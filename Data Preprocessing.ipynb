# Checking for NaN values

```python
df.isna().sum()
```

```python
# search google to populate age where cell is empty

for index, row in df.iterrows():
    if row['age'] is None or pd.isna(row['age']):
        search_name = row['name']
        # Search online for age information
        search_query = f"{search_name} age"
        response = rq.get(f"https://www.google.com/search?q={search_query}")
        soup = bs(response.text, 'html.parser')

        # Extract age from the search results
        age_element = soup.find('div', {'class': 'BNeawe iBp4i AP7Wnd'})
        if age_element:
            found_age = age_element.text
            df.at[index, 'age'] = found_age
```
